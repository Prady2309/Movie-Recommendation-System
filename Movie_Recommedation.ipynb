# Feature Conversion

```python
# To extract only 'name' from the "genres" and "keywords"
import ast
def convert(obj):
    val = []
    for vname in ast.literal_eval(obj):
        val.append(vname['name'])
    return val
```

```python
new_df['genres'] = new_df['genres'].apply(convert)
new_df['keywords'] = new_df['keywords'].apply(convert)
```

```python
new_df.iloc[3].crew
```

```python
# extract top 3 "cast" members and convert to readable format
def cast_convert(obj):
    cast_name = []
    counter = 0
    for name in ast.literal_eval(obj):
        if counter != 3:
            cast_name.append(name['name'])
            counter += 1
        else:
            break
    return cast_name

new_df['cast'] = new_df['cast'].apply(cast_convert)
```

```python
# From "crew", extracting only "Director" name
def director_name(obj):
    name = []
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            name.append(i['name'])
            break
    return name
```

```python
new_df['crew'] = new_df['crew'].apply(director_name)
```

```python
new_df.head(10)
```

```python
new_df['overview'][0]
```

```python
new_df['overview'] = new_df['overview'].apply(lambda x:x.split())
new_df.head(10)
```

```python
# Removing spaces in between words/after punctuation marks etc. 
new_df['crew'] = new_df['crew'].apply(lambda x:[i.replace(" ", "") for i in x])
new_df['keywords'] = new_df['keywords'].apply(lambda x:[i.replace(" ", "") for i in x])
new_df['genres'] = new_df['genres'].apply(lambda x:[i.replace(" ", "") for i in x])
new_df['cast'] = new_df['cast'].apply(lambda x:[i.replace(" ", "") for i in x])
```

```python
new_df.head()
```

```python
# Create a new column "tag" by merging all the columns
new_df['tags'] = new_df['overview'] + new_df['genres'] + new_df['keywords'] + new_df['cast'] + new_df['crew']
new_df.head(10)
```

```python
final_df = new_df[['movie_id', 'title', 'tags']]
final_df.head(10)
```

```python
final_df.iloc[0].tags
```

```python
final_df['tags'] = final_df['tags'].apply(lambda x: ' '.join(x))
```

```python
final_df['tags'][0]
```

```python
final_df['tags'] = final_df['tags'].apply(lambda x:x.lower())
final_df
```

# Feature Extraction

```python
# Shows the vectors of each words based on frequency
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000, stop_words='english')   # 5000 words limit
```

```python
cv.fit_transform(final_df['tags']).toarray().shape    # 4806 -> movies(rows) and 5000 -> words(cols)
```

```python
vectors = cv.fit_transform(final_df['tags']).toarray()
```

```python
vectors[0]
```

```python
len(cv.get_feature_names())
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    t = []
    for i in text.split():
        t.append(ps.stem(i))  # stemming individual words to their root form
    return " ".join(t)

final_df['tags'] = final_df['tags'].apply(stem)
```

# Cosine Similarity
measures the document similarity in text analysis by checking if two vector are pointing in the same direction

```python
from sklearn.metrics.pairwise import cosine_similarity    # finds the angle between vectors (not distance)
```

```python
similar = cosine_similarity(vectors)   # dist inversely proportional to similarity
similar.shape
```

```python
print(vectors[0])
print(similar[0])
```

```python
# creating enumerate object of sorted values
sorted(list(enumerate(similar[0])), reverse=True, key=lambda x:x[1])[1:6]
```

# Recommendation

```python
def recommend(movie):
    movie_index = final_df[final_df['title']==movie].index[0] # Search for the title and get the 1st match
    distances = similar[movie_index]  # similarity scores between the given movie and all other movies.
    movie_list = sorted(list(enumerate(distances)), reverse=True, key=lambda x:x[1])[1:6]

    for i in movie_list:
        print(final_df.iloc[i[0]].title)        # i[0] -> index
```
